```python
import numpy as np  # numpy 패키지 가져오기
import matplotlib.pyplot as plt # 시각화 패키지 가져오기
import seaborn as sns # 시각화

## 2.데이터 가져오기
import pandas as pd # csv -> dataframe으로 전환

## 3.데이터 전처리
from sklearn.compose import ColumnTransformer#데이터 컬럼별로 다른 전처리 방식
from sklearn.pipeline import Pipeline#순차적 자동화
from sklearn.impute import SimpleImputer#누락된 값
from sklearn.preprocessing import StandardScaler, OneHotEncoder#표준화,가변수처리
from sklearn.preprocessing import LabelEncoder#변주형 데이터를 숫자형으로 변환

# 4. 훈련/검증용 데이터 분리
from sklearn.model_selection import train_test_split

## 5.모델구축
# from sklearn import linear_model
from sklearn.linear_model import Lasso     # 파라미터 튜닝을 위해 lasso 모델로 변경
#과적합 방지

## 6.모델검정
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 7.최적화
from sklearn.model_selection import GridSearchCV # 하이퍼파라미터 튜닝
```

```python
housing_df['주거유형']=housing_df['주거유형'].replace([1,2,3,4,5],['단독주택','2가구변경','듀플렉스','타운젠트바깥쪽','타운젠트안쪽'])
housing_df['판매유형'] = housing_df['판매유형'].replace ([1,2], ['보증증서','법원관리증서'])
housing_df['판매조건'] = housing_df['판매조건'].replace ([1,2], ['정상판매','압류및공매도'])
housing_df.head()
```

```python
#data: X target:y로 분리
X=housing_df.drop(['가격'],axis=1)
X.head()
```

```python
#target(y)추출,y값을 series에서 배열로 변환
y = housing_df['가격']
y.head()
```

```python
y=np.array(y)#배열 변환
```

```python
y=StandardScaler().fit_transform(y[:,np.newaxis]).flatten()
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               test_size=0.3,
                                               random_state=1)
```

```python
#pipeline으로 세팅
#수치형 변수(y는 제외)
numeric_features=['연면적','품질','상태','건축년도','리모델링년도',
                  '지하면적','차고면적','면적_1층','면적_2층']
numeric_transformer=Pipeline(
    steps=[('imputer',SimpleImputer(strategy='mean')),#누락된 값은 중앙값으로 처리
           ('scaler',StandardScaler())#표준화
          ]
)

```

```python
#범주형 변수(y는 제외)
catagorical_features=['주거유형','판매유형','판매조건']
categorical_transformer=Pipeline(
    steps=[('imputer',SimpleImputer(strategy='constant',
                                    fill_value='missing')),#누락된 값은 missing으로 상수 처리
           ('encoder',OneHotEncoder(handle_unknown='ignore'))#가변수처리 및 알 수 없는 메시지 무시
           ]
)
```

```python
#preprocessor로 데이터 변환 세팅
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
ColumnTransformer(transformers=[('num',
		                                 Pipeline(steps=[('imputer', SimpleImputer()),
		                                                 ('scaler', StandardScaler())]),
		                                 ['연면적', '품질', '상태', '건축년도', '리모델링년도', '지하면적',
		                                  '차고면적', '면적_1층', '면적_2층']),
		                                ('cat',
		                                 Pipeline(steps=[('imputer',
		                                                  SimpleImputer(fill_value='missing',
		                                                                strategy='constant')),
		                                                 ('encoder',
		                                                  OneHotEncoder(handle_unknown='ignore'))]),
		                                 ['주거유형', '판매유형', '판매조건'])])
```

```python
pd.DataFrame(preprocessor.fit_transform(X_train)).head()
```

```python
Ir_pipe=Pipeline(#Ir:LinearRegression 약자
    steps=[
       ("preprocessor",preprocessor),
       ("Ir_reg",Lasso())
       ]
)
```

```python
Ir_pipe.get_params().keys()#수정할 수 있는 파라미터 확인
```

```python
Ir_grid={
    "Ir_reg__tol":np.logspace(-5,-1,5),
    "Ir_reg__alpha":np.logspace(-5,-1,5),
}
Ir_grid
```

```python
Ir_gs=GridSearchCV(estimator=Ir_pipe,
                   param_grid=Ir_grid,
                   scoring='r2', #mse,r2
                   cv=10,
                   n_jobs=-1,
                   verbose=10,#프로세스 진행사항 확인
                   return_train_score=True #훈련결과 저장
                  )
```

```python
Ir_gs.fit(X_train,y_train)
```

```python
Ir_gs.cv_results_#결과확인r_gs.cv_results_#결과확인
```

```python
#분석결과를 표로 정리
results_df=pd.DataFrame(Ir_gs.cv_results_)
results_df=results_df.sort_values(by=["rank_test_score"])
results_df[["params","rank_test_score","mean_test_score","std_test_score"]]
```
